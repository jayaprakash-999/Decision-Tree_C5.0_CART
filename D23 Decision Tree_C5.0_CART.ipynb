### Building Decision Tree Classifier using Entropy Criteria

```python

print('x_train',x_train.shape)
print('x_test',x_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

```python
x_train.head()
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)


""" DecisionTreeClassifier is a class --> from the sklearn.tree module,
    which is used for building decision tree models """

""" criterion  --> Make a decision or form an opinion about somebody
    criterion = 'entropy': This parameter specifies that the model should use "entropy"
    as the criterion for splitting the data at each node of the tree
"""
# max_depth=3 ---> 3 levels of tree
```

```python
#PLot the decision tree
tree.plot_tree(model);
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(model,
               feature_names = fn,
               class_names=cn,
               filled = True);

## feature names
## class_names  , category name
## dpi=300 determines the resolution of the plot (dots per inch). Higher dpi means a higher-quality image


# nrows = 1, ncols = 1: This specifies that we want a single plot
# model: This is the trained decision tree model that we want to visualize.
# This argument provides the feature names (fn) to be displayed in the tree nodes.
# This argument supplies the class names (cn) to be shown in the tree nodes

# filled = True: This option fills the nodes of the tree with colors representing the majority class in each node, making the visualization more informative.
# values output x
```

```python
#Predicting on test data
preds = model.predict(x_test)     # predicting on test data set
pd.Series(preds).value_counts()   # getting the count of each category

# Accuracy = (Number of Correct Predictions) / (Total Number of Predictions)
# Accuracy = 90 / 100 = 0.9 or 90%
```

```python
pd.crosstab(y_test,preds)

# getting the 2 way table to understand the correct and wrong predictions
```

```python
# getting the 2 way table to understand the correct and wrong predictions
print(classification_report(y_test,preds))
```

```python
# Accuracy
np.mean(preds==y_test)
```

#### Decision Tree Regression Example

```python
# Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
```

```python
array = iris.values
X = array[:,0:3]
y = array[:,3]
```

```python
X
```

```python
X[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

# train_size = 67
# test_size  = 33
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
#Find the accuracy
model.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score

# Predict the target values using the trained model and the test features
y_pred = model.predict(X_test)

# Calculate the R-squared score
r2_score(y_test,y_pred)
```
